```python
import pandas as pd
import torch
import torch.nn as nn #neural networks
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
class MLPClassifier(nn.Module):

    def __init__(self,input_size, hidden_size, output_size):
        super(MLPClassifier,self).__init__()
        self.input_layer = nn.Linear(input_size,hidden_size)  # classe Linear implementa uma fully connected layer
        self.hidden_layer = nn.Linear(hidden_size,output_size)
        self.relu = nn.ReLU()
        self.output_layer = nn.LogSoftmax(dim=1)

    def forward(self,x): # x eh a observacao da entrada
        x = self.relu(self.input_layer(x))
        x = self.relu(self.hidden_layer(x))
        x = self.output_layer(x)
```
